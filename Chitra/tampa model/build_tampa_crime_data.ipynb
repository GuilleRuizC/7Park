```python
acs_zip_codes_2015_5yr["Year"]=2015
acs_zip_codes_2016_5yr["Year"]=2016
acs_zip_codes_2017_5yr["Year"]=2017
acs_cols_2015=acs_zip_codes_2015_5yr.columns
acs_cols_2016=acs_zip_codes_2016_5yr.columns
acs_cols_2017=acs_zip_codes_2017_5yr.columns
common_columns=list((((
(set(acs_cols_2015))).intersection(set(acs_cols_2016)))
                    .intersection(set(acs_cols_2017))))
```

```python
frames=[acs_zip_codes_2015_5yr[common_columns],
        acs_zip_codes_2016_5yr[common_columns],
        acs_zip_codes_2017_5yr[common_columns]]
acs_all_tampa=pd.concat(frames)
```

```python
zipcode_pop=acs_all_tampa.groupby(["geo_id","Year"])["total_pop"].agg("sum").reset_index()
zipcode_pop.columns=["zipcode","Year","population"]
```

```python
zipcode_pop["zipcode"]=zipcode_pop["zipcode"].astype(str)
```

```python
crime_2015=pd.read_csv("florida_crimes_2015.csv")
crime_2016=pd.read_csv("florida_crimes_2016.csv")
crime_2017=pd.read_csv("florida_crimes_2017.csv")
```

```python
crime_2015.loc[:,'Total Index Crimes']=crime_2015['Total Index Crimes'].str.replace(",","").str.replace("DNR","0").astype(float)
crime_2016.loc[:,'Total Index Crimes']=crime_2016['Total Index Crimes'].str.replace(",","").str.replace("DNR","0").astype(float)
crime_2017.loc[:,'Total Index Crimes']=crime_2017['Total Index Crimes'].str.replace(",","").str.replace("DNR","0").astype(float)
```

```python
missingMetros = pd.isnull(zriall["Metro"])
zriall=zriall[~missingMetros]
zriall.rename(columns={"RegionName":"zipcode"},inplace=True)
```

```python
tampa_metro = zriall[zriall.Metro.str.contains('Tampa')]
tampa_metro["zipcode"]=tampa_metro["zipcode"].astype(str)
```

```python
tampa_zipcode_pop=pd.merge(zipcode_pop,tampa_metro, on="zipcode")
```

```python
tampa_zipcode_pop_county=tampa_zipcode_pop[['zipcode','Year','population','CountyName']]
```

```python
tampa_zipcode_pop_county
```

```python
tampa_zipcode_pop_county["population"]=tampa_zipcode_pop_county["population"].astype(float)
tampa_code_pop_total_county=tampa_zipcode_pop_county.groupby("CountyName")["population"].agg("sum").reset_index()
```

```python
tampa_zipcode_pop_county_perc=pd.merge(tampa_zipcode_pop_county,tampa_code_pop_total_county,on="CountyName")
```

```python
tampa_zipcode_pop_county_perc
```

```python
def getCrimeData(row):
    if (row["Year"]==2015):
        county=row["CountyName"].split(" ")[0].upper()
        county_crime=crime_2015[crime_2015["County"].str.contains(county)]['Total Index Crimes'].sum()
        return round(county_crime*(row['population_x']/row['population_y']))
    elif (row["Year"]==2016):
        county=row["CountyName"].split(" ")[0].upper()
        county_crime=crime_2016[crime_2016["County"].str.contains(county)]['Total Index Crimes'].sum()
        return round(county_crime*(row['population_x']/row['population_y']))
    elif (row["Year"]==2017):
        county=row["CountyName"].split(" ")[0].upper()
        county_crime=crime_2017[crime_2017["County"].str.contains(county)]['Total Index Crimes'].sum()
        return round(county_crime*(row['population_x']/row['population_y']))
```

```python
tampa_zipcode_pop_county_perc.loc[:,"crime"]=tampa_zipcode_pop_county_perc.apply(getCrimeData,axis=1)
```

```python
tampa_zipcode_pop_county_perc[["zipcode","Year","CountyName","crime"]].to_csv("tampa_crime_2015_2017.csv")
```
